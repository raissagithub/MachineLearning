```python
# import numpy
import numpy as np
# numpy allows to execute operands on arrays and matrices
import matplotlib.pyplot as plt
```

## 1. Sigmoid function

### Formula of sigmoid:  $ \sigma(x) = \frac{1}{ 1+ e^{-x}} $


```python
# The sigmoid function

def sigmoid(x):
    """
    This function computes the sigmoid of 'x',

    where 'x' => A scalar or numpy array,

    and returns 's' => sigmoid(x)
    """    

    s = 1 / (1+np.exp(-x))

    
    return s
```

```python
# What does the sigmoid function look like?
x = np.arange(-10., 10., 0.1)
plt.plot(x,sigmoid(x), color="magenta")
plt.text(-9, 0.9, r'$sigmoid(x)=\frac{1}{1+e^{-x}}$', fontsize=12)
plt.axvline(x=0, color="black")
plt.show()
```

## 2. Derivative of sigmoid function
### Formula of sigmoid derivative: $sigmoid\_derivative(x) = \sigma(x) (1 - \sigma(x))$


```python
# The derivative of the sigmoid function

def sigmoid_d(x):
    """
    This function computes the derivative (i.e. gradient) of the sigmoid function, taking 'x' as the input.
    
    where 'x' => a scalar or numpy array

    and returns
    ds => the derivative of the sigmoid function.
    """
    
    s = 1/ (1+np.exp(-x))
    ds = s* (1-s)

    
    return ds
```

```python
# What does the derivative of the sigmoid function look like?
x = np.arange(-10., 10., 0.1)
plt.plot(x,sigmoid_d(x), color="green")
plt.ylim(top=0.3) 
plt.text(-10, 0.28, r'$sigmoid\_d(x) = \sigma(x) (1 - \sigma(x))$', fontsize=12)
#plt.axvline(x=0, color="black")
plt.show()
```
